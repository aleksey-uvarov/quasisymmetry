```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../h4_square_quasi_symmetry_comm_optimization.csv")
```

```python
df.columns
```

```python
for i, row in df.iterrows():
    to_print = "{0:1.3f} & {1:1.4f} & {2:1.4f} & ".format(
        row["Geometry_Param"], row["V_0"], row["V_optimized"])

    comm_reduction = row["Sum_CommSq_0"] / row["Sum_CommSq_Optimized"]

    to_print += "{0:1.5e} & {1:1.5e} & {2:1.4f} & ".format(
        row["Sum_CommSq_0"], row["Sum_CommSq_Optimized"], comm_reduction)

    to_print += r"{0:1.1f} : {1:1.1f} : {2:1.3f} \\".format(
        row["a_opt"], row["b_opt"], row["c_opt"])
    print(to_print)
```
